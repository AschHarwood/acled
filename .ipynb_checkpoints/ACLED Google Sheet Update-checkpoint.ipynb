# How to update data to Google Sheet from ACLEd API

1. Download most recent data from ACLED
2. Remove unwanted countries (Egypt, Libya, Morroco, Tunisia, and Algeria)
3. Grab data_id from google sheet
4. Compare to event_id in new data, and remove any duplicates
5. Write remaining data to sheet
6. 

## Remove unwanted countries from ACLED

```python
remove_country_list = ['Egypt', 'Libya', 'Morocco', 'Algeria', 'Tunisia']
```

```python
filtered_list = []
for item in data:
    if item['country'] not in remove_country_list:
        filtered_list.append(item)
```

## Grab data_id from Google Sheet

```python
scope = ['https://spreadsheets.google.com/feeds']
credentials = ServiceAccountCredentials.from_json_keyfile_name('ACLED_to_Google_Sheets-e244e52c2f43.json', scope)
gc = gspread.authorize(credentials)
sheet = gc.open('ACLED_Master').sheet1
```

```python
#saves event_id to values_list
values_list = sheet.col_values(2)
```

```python
#compare values_list with filtered_list and removes dupes
#returns deduped filtered list
for item in filtered_list:
    if item['event_id_cnty'] in values_list:
        filtered_list.remove(item)
```

## Update Google Sheet

```python
df = pd.DataFrame(filtered_list)
```

```python
column_list_api = df.columns.tolist()
```

#need to rearrange column list, since df columns are different than columns in google sheet
1. Convert row_list from Google Sheet to lower case
2. Convert row_list to unicode
3. use row_list to rearrange columsn for upload

```python
row_list_from_sheets = sheet.row_values(1)
```

```python
#makes row_list lower case
row_list_lower = []
for i in row_list_from_sheets:
    row_list_lower.append(i.lower())
    
```

```python
#makes row_list unicode
row_list_unicode = []
for i in row_list_lower:
    row_list_unicode.append(i.decode('utf-8'))
```

```python
#removes those weird spots at the bottom
row_list_unicode.pop()
row_list_unicode.pop()
```

```python
#initiates a reorganized dataframe
df_reorg = df[row_list_unicode]
```
