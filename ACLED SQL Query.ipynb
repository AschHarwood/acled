```python
import pandas as pd
import requests
import matplotlib.pyplot as plt
import sqlite3 as lite
import sqlalchemy
%matplotlib inline
```

```python
r = requests.get('http://acleddata.com/api/acled/read')
```

```python
data = r.json()
data_set = data['data']
```

```python
df = pd.DataFrame(data_set)
```

```python
con = lite.connect('acled4.db')
```

```python
df.to_sql('acled', con, if_exists='append', index=False)
```

```python
cur = con.cursor()
```

```python
cur.execute('SELECT * FROM acled')
all_rows = cur.fetchall()
```
